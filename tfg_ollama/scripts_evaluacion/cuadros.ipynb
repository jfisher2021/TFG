```python
import numpy as np 
import os
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv("../cuadros.csv")
dataset_frame = pd.DataFrame(dataset)
# pd.set_option('display.max_columns', len(dataset.columns))

# Visualizar las primeras filas
print("Primeras filas del conjunto de datos:")
print(dataset_frame.head())

# Mostrar los nombres de las columnas
print("\nNombres de las columnas:")
print(dataset_frame.columns)
```

```python
cuadros_franceses = dataset_frame[dataset_frame.Pais_del_Autor == "Francia"]
cuadros_picasso = dataset_frame[dataset_frame.Autor == "Pablo Picasso"]
cuadros_franceses_y_picasso = pd.concat([cuadros_franceses, cuadros_picasso]).drop_duplicates().reset_index(drop=True)

print("Cuadros franceses y los de Picasso:")
print(cuadros_franceses_y_picasso[['Nombre', 'Autor', 'Pais_del_Autor', 'Año']])
```
